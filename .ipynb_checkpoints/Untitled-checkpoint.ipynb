```python
X=dataset[['maxtemp','evaporation_final']]
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)
pca.components_
```

```python
pca.components_[0]
#0.93779246  of the first column (X[:,0]) and  0.34719635 of the second dimension (X[:, 1])
```

```python
pca.components_[1]
```

```python
pca.explained_variance_
```

```python
print(pca.explained_variance_ratio_)

#For this case, we observe that only 1 component explains ~89% of the variance (information) in our data. It means that you can reduce the number of dimensions to 1, and it will still retain ~89% of its signal.
```

```python
pca = PCA(n_components=1)
pca.fit(X)
X_pca = pca.transform(X)
print("original shape:   ", X.shape)
print("transformed shape:", X_pca.shape)
pca.explained_variance_ratio_
```
